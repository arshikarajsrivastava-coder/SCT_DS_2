```python
import pandas as pd
df = pd.read_csv("C:/Users/ansh_/Documents/titanic/train.csv") 
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode())
print(df.describe())  # summary of numeric columns
print(df['Sex'].value_counts())   # gender distribution
print(df['Pclass'].value_counts())  # class distribution
print(df['Survived'].value_counts()) # survival stats
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Survival count
sns.countplot(x='Survived', data=df)
plt.title('Survival Counts')
plt.show()

# Survival by sex
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.show()

# Survival by class
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution')
plt.show()

df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 12, 18, 35, 60, 80], labels=['Child', 'Teen', 'Young Adult', 'Adult', 'Senior'])
sns.barplot(x='AgeGroup', y='Survived', data=df)
plt.title('Survival Rate by Age Group')
plt.show()
```
